```python
%matplotlib inline
```

```python
#Import Dependencies 
import matplotlib.pyplot as plt
```

```python
# Set x-axis for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug","Sept", "Oct", "Nov","Dec"]
y_axis = [10.02,23.24,39.20,35.42,32.34,27.04,43.82,10.56,11.85,27.90,20.71,20.09]
```

```python
#Create Plot
plt.plot(x_axis,y_axis)
```

```python
#Create with ax.plt
fig, ax= plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
#create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
#Create the plot
plt.plot(x_axis,y_axis)
plt.show()
```

```python
    #Create the Plot and add Labels
plt.plot(x_axis,y_axis, label = "Boston")
    #Create X,y labels
plt.xlabel("Date")
plt.ylabel("Fare($)")
    #set Y limit 0-45
plt.ylim(0,45)
    #create a title
plt.title("Pyber Fare by Month")
    #add Legend
plt.legend()
```

```python
    #Create the Plot and add Labels
plt.plot(x_axis,y_axis, marker="*", color = "green", linewidth=2, label= "Boston")
    #Create X,y labels
plt.xlabel("Date")
plt.ylabel("Fare($)")
    #set Y limit 0-45
plt.ylim(0,45)
    #create a title
plt.title("Pyber Fare by Month")
    #add Legend
    #add grid
    plt.grid()
plt.legend()
```

```python
#Create bar graph
plt.bar(x_axis,y_axis)
```

```python
#annotate bar chart
plt.bar(x_axis,y_axis, color= "green", label="Boston")
plt.xlabel("Date")
plt.ylabel("fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Horizontal bar chart
plt.barh(x_axis,y_axis, color="pink")
plt.gca().invert_yaxis()
```

```python
#create a plot with ax.plt()
fig, ax =plt.subplots()
ax.barh(x_axis,y_axis)
```

```python
plt.plot(x_axis,y_axis, "o")
```

```python
y_axis_larger= []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis,y_axis, s=y_axis_larger)
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python


explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.subplots(figsize=(8, 8))
plt.show
```

```python
# Add color for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, 
    explode=explode_values, 
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%')
#plt.subplots(figsize=(8, 8))
plt.show
```
