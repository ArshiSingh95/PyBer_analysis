```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
names = ['Tony', 'Joe', 'Sally', 'Andrea']
num_marbles = [100, 500, 150, 20]
```

```python
plt.scatter(names, num_marbles,color="skyblue", s=num_marbles)
plt.xlabel("names")
plt.ylabel(" # of Marbles")
plt.title("Bar")
```

```python
plt.bar(names, num_marbles,color="skyblue")
plt.xlabel("names")
plt.ylabel(" # of Marbles")
plt.title("Bar")
```

```python

explode_values = (0.1, 0.4, 0.2, 0.2, )
plt.pie( num_marbles, explode=explode_values,  labels=names, autopct='%.1f%%', )
plt.xlabel("names")
plt.ylabel(" # of Marbles")
plt.title("Pie")
```
